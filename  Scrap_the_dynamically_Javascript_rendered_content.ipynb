```python
from selenium import webdriver
import pandas as pd
# pandas to store data in csv and manipulate data 
```

```python
url ='https://techolution.app.param.ai/jobs/'
```

```python
# web browser engine, geckodriver path
driver = webdriver.Firefox('/usr/local/bin')
```

```python
driver.get(url)
```

```python
cat = driver.find_elements_by_xpath("//*[@class='sixteen wide mobile eleven wide tablet eleven wide computer column']")
cat
```

```python
#loop and save data we need 
for result in cat:
    job_name = result.text
job_name
```

```python
# for col
results = driver.find_elements_by_xpath("//*[@class='ui stackable grid container']")
for result in results:
    job_col = result.text
col=job_col.split('Information Technology\n')[0].split('\n')
col
```

```python
#new col
col =['Categories',
    'Job Openings',
 'Job types',
 'Locations',
 'Experience',
 'Post_date',]
```

```python
# split jobs
joblist=job_name.split('ago\n')
```

```python
joblist=joblist[:27]
joblist
```

```python
df = pd.DataFrame(data=joblist,columns=['job'])
```

```python
df['job']=df['job'].apply(lambda x : x.replace('·','\n').split('\n'))
total_len=len(df.loc[0][0])
total_len
```

```python
total_len=len(df.loc[0][0])
total_len
```

```python
new_df= pd.DataFrame(columns=col)
df.loc[0].values[0][0]
```

```python
# convert to diff types
for i in range(len(df)):
        if(total_len==len(df.loc[i][0])):
            #if Categories
            j=0
            #store categoriaes
            cat=[df.loc[i][0][0]]
           # print(j)
            new_df.loc[i] =cat+[df.loc[i][0][1-j]]+[df.loc[i][0][2-j]]+[df.loc[i][0][3-j]]+[df.loc[i][0][4-j]]+[df.loc[i][0][5-j]]
        else:
            j=1
            new_df.loc[i] =cat+[df.loc[i][0][1-j]]+[df.loc[i][0][2-j]]+[df.loc[i][0][3-j]]+[df.loc[i][0][4-j]]+[df.loc[i][0][5-j]]
            
```

```python
new_df
```

```python
# data according to the date of posting(Most old first)
def covert(x):
    v=x.split(' ')
    if(v[0]=='a'):
        r=30 #appox
    elif(v[1]=='hours'):
        r=int(v[0])*1/24
    elif(v[1]=='months'):
        r=int(v[0])*30
    else:
        r=int(v[0])*1
    return r
```

```python
new_df['r']=new_df['Post_date'].apply(lambda x: covert(x))
new_df=new_df.sort_values(by='r',ascending=False)
```

```python
new_df.drop(['r'],inplace=True,axis=1)
new_df.to_csv('date_of_posting',index=False)
```
